```python
# 导入定点数核心模块（无需修改）
import fixNum_final

# ======================
# 完全参考 v2 写法：独立菜单函数
# ======================
def show_menu():
    print("\n===== Fixed Point Number Menu =====")
    print("1. Multiply (Task 3)")
    print("2. Power (Integer Exponent)")
    print("3. Power (Fixed Point Exponent)")
    print("4. Change Precision")
    print("0. Exit")
    print("===================================")

# ======================
# 完全参考 v2 写法：检查是否为0
# ======================
def check_zero(val):
    return val.a == 0 and val.b == 0

# ======================
# 完全参考 v2 写法：主程序入口
# ======================
def main():
    print("Fixed Point Number Calculator (Final Version)")
    print(f"Current Precision: {fixNum_final.prec} decimal places")
    
    while True:
        show_menu()
        choice = input("Enter your choice: ")

        # 1. 乘法 (Task 3)
        if choice == '1':
            print("\n[Multiply Two Fixed Point Numbers]")
            num1 = fixNum_final.fixNum()
            num1.input()
            
            num2 = fixNum_final.fixNum()
            num2.input()
            
            if check_zero(num1) or check_zero(num2):
                print("Error: Fixed-point number is zero! Do nothing.")
                continue
            
            result = num1 * num2
            print(f"Result: {num1} * {num2} = {result}")

        # 2. 整数次幂 (Distinction)
        elif choice == '2':
            print("\n[Power: Fixed ^ Integer]")
            base = fixNum_final.fixNum()
            base.input()
            
            if check_zero(base):
                print("Error: Fixed-point number is zero! Do nothing.")
                continue
            
            try:
                exp = int(input("Enter integer exponent: "))
                result = base ** exp
                if result:
                    print(f"Result: {base} ^ {exp} = {result}")
            except:
                print("Invalid input!")

        # 3. 定点数次幂 (Distinction)
        elif choice == '3':
            print("\n[Power: Fixed ^ Fixed]")
            base = fixNum_final.fixNum()
            base.input()
            
            if check_zero(base):
                print("Error: Fixed-point number is zero! Do nothing.")
                continue
            
            exp = fixNum_final.fixNum()
            exp.input()
            result = base ** exp
            if result:
                print(f"Result: {base} ^ {exp} = {result}")

        # 4. 修改精度
        elif choice == '4':
            try:
                new_prec = int(input("Enter new precision (>0): "))
                if new_prec > 0:
                    fixNum_final.prec = new_prec
                    print(f"Precision changed to: {new_prec}")
                else:
                    print("Precision must > 0")
            except:
                print("Invalid input!")

        # 0. 退出
        elif choice == '0':
            print("Program closed.")
            break

        # 无效选项
        else:
            print("Invalid choice! Try again.")

# ======================
# 完全参考 v2 写法：运行主函数
# ======================
if __name__ == "__main__":
    main()
```
